```python
#importing the dependencies/modules
import requests
import csv
from bs4 import BeautifulSoup as soup
import pandas as pd

#loding the HTML and parsing the HTML 

link = requests.get("https://www.zaubacorp.com/company/BUSINESSONBOT-PRIVATE-LIMITED/U72900KA2020PTC136387/")
soup = soup(link.content, 'html.parser')

#creating the csv 

file = "ynos_1.csv"
open(file,"w")
csv_writer = csv.writer(open(file, 'r+', encoding = "utf-8"))

```

```python

# tr - Table Row
# th - Table Header
# tr - Data cell of table

# Each Table is being iterated and the neccessary elements of the table are taken and are written to the csv file

for tr in soup.find_all("tr"):
    data = []

    for th in tr.find_all("th"):
        data.append(th.text)
    if data:
        print("Inserting headers : {}".format(','.join(data)))
        csv_writer.writerow(data)
        continue

# td.a - anchor element

    for td in tr.find_all("td"):
        if td.a:
            data.append(td.a.text.strip())
        else:
            data.append(td.text.strip())
    if data:
        print("Inserting data: {}".format(','.join(data)))
        csv_writer.writerow(data)
        
```
